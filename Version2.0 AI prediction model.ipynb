```python
#1
#Ignore warnings
import warnings; 
warnings.simplefilter('ignore')

#2
#Install and import Dependencies
import pandas as pd
from fbprophet import Prophet
#3
#Read the data 
fileName = 'Processed week Multiplied Transfer Station tpd (2.5).csv'

df = pd.read_csv(fileName)
#df.head() #Check if the data are correctly taken
#df.dtypes #Check if the data match with the requirement of subsequent functions


#4
df2 = pd.DataFrame()
df2['ds'] = df[['date']]
df2.tail()

df2['y'] = df[['IETS']]
tempString = 'IETS'

print(df2)

#6
#Train the model
m = Prophet(interval_width=0.95, daily_seasonality=True)
model = m.fit(df2)

#7
# Make prediction over the future
future = m.make_future_dataframe(periods=156,freq='W') ######## w could be 'day' or 'week'
forecast = m.predict(future)
forecast.head()

#8
#Plot the prediction result-1
plot1 = m.plot(forecast)
#Set the parameters of the Plot
p = plot1.gca()
p.set_title("Prediction of the amount of MSW disposed of at landfills", size=20)
p.set_xlabel("Year", size=14)
p.set_ylabel("MSW (Tons)", size=14)
p.tick_params(axis="x", labelsize=14)
p.tick_params(axis="y", labelsize=14)
#exporting the result
# info = pd.DataFrame(forecast)   
# info.to_csv('prediction.csv')

#10
# resultOutput = 'Result of NWNTTS of '+fileName[9:] 
resultOutput = 'Result of ' +tempString +' of '+fileName[9:] 
forecast.to_csv(resultOutput)
```
